#### `capital` column : Nan values but written as "—"  in [Baker Island, Coral Sea Islands, Howland Island, Jarvis Island, Johnston Atoll, Kingman Reef, Midway Atoll, Palmyra Atoll]

##### Define
search for there capitals if they have then zip it with countries names other leave it as NaN

##### Code

```python
countries = ['Baker Island','Coral Sea Islands', 'Howland Island', 'Jarvis Island', 'Johnston Atoll','Kingman Reef', 'Midway Atoll', 'Palmyra Atoll']
capitals = ['administered from Washington, DC','administered from Canberra', 'administered from Washington, DC','administered from Washington, DC',np.nan,np.nan,np.nan,np.nan]
both = dict(zip(countries,capitals))
for i in both:
    clean.loc[clean['country'] == str(i),('capital')] = both[i]
```

##### Define
take official or political capital

##### Code

```python
countries = ['Tokelau',
           'Côte d’Ivoire (Ivory Coast)',
           'Mayotte', 'Saint Helena, Ascension and Tristan da Cunha',
           'South Africa',
           'Swaziland','Tanzania']
for i in countries:
    print(i , "\tcaptal:\t", clean.loc[clean['country'] == i,('capital')])
    print()
```

```python
capitals = ['Nukunonu','Yamoussoukro','Mamoudzou','Jamestown','Pretoria','Mbabane','Dodoma']

both = dict(zip(countries,capitals))
for i in both:
    clean.loc[clean['country'] == str(i),('capital')] = both[i]
```


##### Test

#### `capital` column: duplicated comma

##### Define
convert colum to list of strings then convert it againg to string but without duplicated comma

##### Code

```python
temp = clean[clean['capital'].str.contains(',,',na = False)]
indices = []
for index, row in temp.iterrows():
    indices.append(index)
    l = [c for c in row.capital.split(',') if c !='']
    s = ','.join(str(w) for w in l)
    temp.loc[temp.country == row.country, ('capital')] = s
clean.loc[clean['capital'].str.contains(',,',na = False)] = temp
```

```python
clean[clean['capital'].str.contains(',,',na = False)]
```

#### `language` column: remove duplicated comma

##### Define
using temp data frame to update values

##### Code

```python
temp = clean[clean['language'].str.contains(',,',na = False)]
indices = []
for index, row in temp.iterrows():
    indices.append(index)
    l = [c for c in row.capital.split(',') if c !='']
    s = ','.join(str(w) for w in l)
    temp.loc[temp.country == row.country, ('language')] = s
clean.loc[clean['language'].str.contains(',,',na = False)] = temp
```
