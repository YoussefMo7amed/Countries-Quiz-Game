```python
import os
import requests as req
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
```

## gathering data from url 

```python
url = 'https://affairscloud.com/europe-continent-countries-capitals-currency-code/'
res = req.get(url)
res
```

```python
with open(os.path.join('Australia.html'), mode = 'wb') as file:
    file.write(res.content)
```

## getting data using Beautiful Soup

getting head

```python
columns = []
with open('Australia.html') as file:
    soup = bs(file)
    table = soup.find('table')
    head = table.find_all('th')
    for col in head:
        columns.append(str(col).strip()[4:-5])
columns
```

```python
columns[0] = 'Country'
```

```python
columns
```

```python
df = pd.DataFrame(columns=columns)
```

```python
df
```

```python
with open('Australia.html') as file:
    soup = bs(file,'lxml')
    table = soup.find('table')
    rows = table.find_all('tr')
    for i in range(1,len(rows)):
        row = rows[i].find_all('td')
        values = []
        for value in row:
            s = str(value).strip()[4:-5]
            s = s.replace('<br/>',', ')
            s = s.replace('\n','')
            values.append(s)
        df = df.append(dict(zip(columns,values)), ignore_index=True)
        
```

```python
df
```

## save data 

```python
df.to_csv('Europe.csv', index= False)
```
