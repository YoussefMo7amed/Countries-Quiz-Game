```python
import os
import requests as req
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
```

```python
asia_url = 'https://affairscloud.com/asia-continent-countries-capitals-currency-with-code/'
res = req.get(asia_url)
res
```

```python
with open(os.path.join('Asia'),mode = 'wb') as file:
    file.write(res.content)
```

```python
os.listdir()
```

```python
col = []
with open('Asia') as file:
    soup = bs(file,'lxml')
    table = soup.find('table')
    head = table.find_all('th')
    for i in head:
        col.append((str(i)[4:-5]).strip())
print(col)
```

```python
df = pd.DataFrame(columns=col)
df
```

```python
with open('Asia') as file:
    soup = bs(file,'lxml')
    table = soup.find('table')
    rows = table.find_all('tr')
    for i in range(1,len(rows)):
        row = rows[i].find_all('td')
        ls = []
        for value in row:
            s = str(value)[4:-5]
            s = s.replace('<br/>', ', ')
            s = s.replace('\n', '')
            ls.append(str(s))
        df = df.append(dict(zip(col,ls)), ignore_index= True)
        
```

```python
df
```

```python
df.to_csv('Asia.csv', index= False)
```
